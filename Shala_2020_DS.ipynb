You have a matrix $X$ of size (10,5) filled with random values. Print the following :
* Minimum value and index where it occured, for each column
* Maximum value and the index where it occured, for each column
* Minimum value and index where it occured, for each row
* Maximum value and index where it occured, for each row
* Minimum value and index where it occured, for the entire matrix
* Maximum value and index where it occured, for the entire matrix

````
np.random.seed(0)
X = np.random.rand(10,5)
print(X)
# Print min and max value of entire matrix with indices
print(np.amin(X))
print(np.argmin(X))
print(np.amax(X))
print(np.argmax(X))
# Print min and max value for each column
print(np.amin(X, axis=0))
print(np.amax(X, axis=0))

# Print min and max value for each row
print(np.amin(X, axis=1))
print(np.amax(X, axis=1))

# Print the respective indices
print(np.argmin(X, axis=0))
print(np.argmin(X, axis=1))
print(np.argmax(X, axis=0))
print(np.argmax(X, axis=1))
````

Find and print the following for the matrix X:

````
# mean of each column
print(np.mean(X, axis=0))
# mean of each row
print(np.mean(X, axis=1))
# Standard deviation of each row and column
print(np.std(X, axis=0))
print(np.std(X, axis=1))
# mean and std of entire matrix
print(np.mean(X))
print(np.std(X))
````

````
# Initial imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv
Read the data and display the first 5 rows

````
#Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv

file_url = "http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv"
df1 = pd.read_csv(file_url)
df1.head()
````

### Preprocessing of the Dataset



---


**1. Convert the categorical values of columns 'sex', 'embarked', 'pclass' into dummy values and add them into titanic_data_new DataFrame**




---



````

df_new = pd.get_dummies(df1[['sex','embarked','pclass']])
df_new.head()
````



---


**2. Fill the missing data in Age and Fare columns with the mean value of the respected column then add these columns to titanic_data_new**


---



````
df1.isna().sum()
df_new[['age','fare']] = df1[['age','fare']].fillna(df1[['age','fare']].mean())
df_new.head()
````

###Normalization


---



1. Normalize the columns which have numerical data



---



````
norm_df = df1[['age','fare']]
res_norm_df = (norm_df - norm_df.mean())/norm_df.std()
res_norm_df.head()
````



---
2. Drop the original age, fare columns
3. Append the normalized columns to the titanic_data_new DataFrame.



---



````
df_new = df_new.drop(labels=['age','fare'],axis=1)
df_new = pd.concat((df_new,res_norm_df),axis=1)
df_new.head()
````

Plot the sine function with domain in $[-3\pi,3\pi]$. Label the axes appropriately. Add title $\sin(x)$. Use latex formatting while adding the title. Set the plot size to (10,10)

````
plt.figure(figsize=(10,10))
x = np.linspace(-3*np.pi , 3*np.pi , 100)
plt.plot(x,np.sin(x))
plt.xlabel('Phase')
plt.ylabel('Magnitude')
plt.title(r'$sin(x)$')
plt.show()
````

````
plt.figure(figsize=(10,10))
x = np.linspace(-2*np.pi,2*np.pi,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.xlabel('Angle')
plt.plot('Magnitude')
plt.plot()
````
